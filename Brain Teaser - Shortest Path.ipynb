```python
import numpy as np
import random as rand
```

```python
#This is the solution that I thought of at first. The principle is to start from and randomly select path until the finish
# We cumulate paths in a list if it is not already in registered. We stop the loop once we have failed 100 attempts to find a new path.
# This solution is not very efficient and as the problem gets more complex it gets impossible to solve

t,start,finish = [[1,4,1],[1,9,1],[1,1,1]], (0,0), (2,2)
#t,start,finish =[[1,9,1],[2,9,1],[2,1,1]], (0,0), (0,2)

shape = (np.shape(t)[0]-1,np.shape(t)[1]-1)
points = []
for i in range(shape[0]+1):
    for a in range(shape[1]+1):
        points.append((i,a))

# This function gives all the valid moves from any position and takes in account previous points that were visited
def valid_moves(position,points,previous):
    all_moves = [(position[0]+1,position[1]),(position[0],position[1]+1),(position[0]-1,position[1]),(position[0],position[1]-1)]
    valid = []
    for i in all_moves:
        if i in points and i not in previous:
            valid.append(i)
    return valid

def cheapest_path(t, start, finish):
    paths = []
    attempts = []
    On = True
    
    # The first while stops once 100 attempts are failed
    while On:
    
        path = []
        previous = []
        position = start
        On2 = True
    
        # This second while stops once the path reaches the finish or if there is no valid moves and finish was not reached
        while On2:
            
            if len(valid_moves(position,points,previous)) != 0:
                move = rand.choice(valid_moves(position,points,previous))
                path += [move]
                previous += [position]
                position = move

                if position == finish:
                    On2 = False
                    if path not in paths:
                        paths +=[path]
                        attempts += [True]
                    else:
                        attempts +=[False]
                else:
                    continue
            else:
                path = []
                previous = []
                position = start
            
        if sum(attempts[len(attempts)-100:len(attempts)])==0:
            On = False
    
    # This retrieves the costs for each path that were found and retrieves the cheapest path for the answer       
    costs = []
    cheapest_path = None
    for i in paths:
        cost = 0
        for a in i:
            cost += np.array(t)[a[0],a[1]]
        costs +=[cost]
        if cost == min(costs):
            cheapest_path = [i,cost]
    
    return cheapest_path
                
cheapest_path(t, start, finish)
```

```python
# This is a second solution that is much more efficient. It is a classical path finding algorithm called dijkstra algorithm

#t,start,finish = [[1,4,1,2],[1,9,3,1],[1,1,1,1],[1,3,1,2]], (0,0), (3,3)
t,start,finish = [[1,4,1],[1,9,1],[1,1,1]], (0,0), (2,2)
        

def valid_moves(position,points):
    all_moves = [(position[0]+1,position[1]),(position[0],position[1]+1),(position[0]-1,position[1]),(position[0],position[1]-1)]
    valid = []
    for i in all_moves:
        if i in points:
            valid.append(i)
    return valid


def cheapest_path(t, start, finish):
    
    shape = (np.shape(t)[0]-1,np.shape(t)[1]-1)
    points = []
    for i in range(shape[0]+1):
        for a in range(shape[1]+1):
            points.append((i,a))
    
    # When we start all the points are unvisisted
    unvisited_nodes = points
    # We'll use this dict to save the cost of visiting each node and update it as we move along the graph   
    shortest_path = {}
    # We'll use this dict to save the shortest known path to a node found so far
    previous_nodes = {}
 
    # We'll use np.infto initialize the "infinity" value of the unvisited nodes except for the start node at 0
    for node in unvisited_nodes:
        shortest_path[node] = np.inf  
    shortest_path[start] = 0
    
    
    # The algorithm executes until we visit all nodes
    while unvisited_nodes:
        
        current_min_node = None
        
        for node in unvisited_nodes:
            if current_min_node == None:
                current_min_node = node
            elif shortest_path[node] < shortest_path[current_min_node]:
                current_min_node = node
                
        # The code block below retrieves the current node's neighbors and updates their distances
        neighbors = valid_moves(current_min_node,points)
        for neighbor in neighbors:
            tentative_value = shortest_path[current_min_node] + np.array(t)[neighbor[0],neighbor[1]] #We use the function to find the
            if tentative_value < shortest_path[neighbor]:
                shortest_path[neighbor] = tentative_value
                # We also update the best path to the current node
                previous_nodes[neighbor] = current_min_node
 
        # After visiting its neighbors, we mark the node as "visited"
        unvisited_nodes.remove(current_min_node)

        
    # We are only interested in the result for the finish coordinate
    # Using the previous node list, we unwind the shortest path from start to finsh
    coordinates = []
    last_node = finish
    while last_node != start:
        coordinates += [previous_nodes[last_node]]
        last_node = previous_nodes[last_node]
    coordinates = coordinates[::-1]+ [finish]
    
            
    return coordinates,shortest_path[finish]

cheapest_path(t, start, finish)
```
