This is an example of how IVC: https://www.ivc-online.com/ can be scrapped with selenium for anyone with an account. 
The process is in two steps: 
1. I make an advanced search to retrieve the links to the each company page that is of interest. (If no criterias are selected as in the example, the full list of links will be retrieved and thus the whole website is scrapped)
2. Using the list of links, selenium will click on each page one by one and scrap its full content to place it in a DataFrame

```python
# Now we are inside the search page let's get name in a little list of all the links taht we care about

Stillon = True
Company_names = []
Links = []

while Stillon is True:
    time.sleep(1.5)
    a = chrome.find_elements_by_class_name("SRcompanyName")
    for i in a: 
        Company_names += [i.text]
        Links += [i.get_attribute('href')]
    
    try: 
        time.sleep(6)
        chrome.find_element_by_id("dnn_ctr459_View_SearchResults_btnNext").click()

    except:
        Stillon = False
        a = chrome.find_elements_by_class_name("SRcompanyName")
        for i in a: 
            Company_names += [i.text]
            Links += [i.get_attribute('href')]
```

```python
# Creating a df just to show the result, (unecessary)
df = pd.DataFrame()
df["Company Names"] = Company_names
df["Links"] = Links
df
```
